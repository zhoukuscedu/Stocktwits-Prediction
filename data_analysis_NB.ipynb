```python
file_name = 'AAPL.txt'
csv_name = 'AAPL.csv'
```

```python
#Features
company = stock_twits_text_parser(file_name)
company_metrics = stock_twits_metrics(company, 3, file_name)

#Labels
company_return = return_over_period_T(csv_name,3)

#Cleaned Data
agg_data = feature_selector(company_metrics, company_return, 'close_return', 25, 1000)
company_data = split_data(agg_data, 0.7, 'text', 'close_return', 'bin')

#Split
X_train = company_data['X_train']
Y_train = company_data['Y_train']

X_test = company_data['X_test']
Y_test = company_data['Y_test']
```

# Naive Bayes - Gaussian Priors

```python
#Fit Model
clf_GNB = GaussianNB()
clf_GNB.fit(X_train, Y_train)

#Predict
pred_train_GNB = clf_GNB.predict(X_train)
pred_test_GNB = clf_GNB.predict(X_test)
```

```python
#Generate Confusion Matrices
confusion_train_GNB = confusion_matrix(Y_train, pred_train_GNB)
confusion_test_GNB = confusion_matrix(Y_test, pred_test_GNB)
```

```python
#Plot Confusion Matrix for Training Data
plot_confusion_matrix(confusion_train_GNB, "Training")
```

```python
#Plot Confusion Matrix for Test Data
plot_confusion_matrix(confusion_test_GNB, "Test")
```

```python
#Calculate Metrics
tableGNB = create_metrics_table(Y_train, pred_train_GNB, Y_test, pred_test_GNB)
tableGNB
```

```python
#Generate ROC Curves for Training
pred_train_prob_GNB = clf_GNB.predict_proba(X_train)[:,1]
FPR_train, TPR_train, Threshold_train = roc_curve(Y_train, pred_train_prob_GNB)

#Plot ROC Curve for Training Data
plot_ROC_curve(FPR_train, TPR_train, "Training")
```

```python
#Generate ROC Curves for Test
pred_test_prob_GNB = clf_GNB.predict_proba(X_test)[:,1]
FPR_test, TPR_test, Threshold_test = roc_curve(Y_test, pred_test_prob_GNB)

#Plot ROC Curve for Test Data
plot_ROC_curve(FPR_test, TPR_test, "Test")
```
